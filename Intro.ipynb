```python
# %% [markdown]
# # AI Impact on Jobs 2030 - Data Visualization & Cleaning
# This notebook explores employee demographics, AI risk, and future demand trends.

# %%
%pip install seaborn
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from IPython.display import display, Markdown

# Set plotting style for clean visuals
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# 1. Load the Dataset
df = pd.read_csv("AI_Impact_on_Jobs_2030.csv")

# Quick look at the first few rows
print("--- FIRST 5 ROWS ---")
display(df.head())

# %%
# 2. Inspect Dataset Structure and Missing Values
print("\n--- DATASET INFO ---")
df.info()

print("\n--- MISSING VALUES PER COLUMN ---")
print(df.isnull().sum())

# %%
# 3. Remove Duplicate Rows
initial_shape = df.shape
df = df.drop_duplicates()
final_shape = df.shape

print("\n--- DUPICATE REMOVAL SUMMARY ---")
print(f"Original dataset shape: {initial_shape}")
print(f"Dataset shape after removing duplicates: {final_shape}")
print(f"Removed {initial_shape[0] - final_shape[0]} duplicate rows.")

# %% [markdown]
# ## 4. Distribution of Hiring Trends (2026)
# Let's see the overall market sentiment by visualizing the current hiring trends.

# %%
plt.figure(figsize=(8, 5))
trend_counts = df["Hiring_Trend_2026"].value_counts()

# Creating a bar chart using matplotlib
plt.bar(
    trend_counts.index,
    trend_counts.values,
    color=["#3BAF50", "#EFC108", "#F45336"],
)
plt.title("Hiring Trends in 2026 Across Roles", fontsize=14, fontweight="bold")
plt.xlabel("Hiring Trend")
plt.ylabel("Number of Records")
plt.tight_layout()
plt.show()
markdown_text=""" we can observe that hiring trends are either growing or stable , because they are on the same level, but declining hiring trends are slightly less.
"""
display(Markdown(markdown_text))



# %% [markdown]
# ## 5. AI Replacement Risk vs. Future Demand Score
# Is high demand correlated with high risk? Let's plot a scatter plot with a trendline.

# %%
plt.figure(figsize=(10, 6))

# Clean any potential missing values in these specific columns for polyfit
sample_df = df.dropna(subset=["AI_Replacement_Risk", "Future_Demand_Score"])
sample_df = sample_df.sample(n=min(1000, len(sample_df)), random_state=42)

# Scatter plot
plt.scatter(
    sample_df["AI_Replacement_Risk"],
    sample_df["Future_Demand_Score"],
    alpha=0.5,
    c="#2190F3",
    edgecolors="w",
)


# Adding a trendline using NumPy
m, b = np.polyfit(
    sample_df["AI_Replacement_Risk"], sample_df["Future_Demand_Score"], 1
)
plt.plot(
    sample_df["AI_Replacement_Risk"],
    m * sample_df["AI_Replacement_Risk"] + b,
    color="red",
    linewidth=2,
    label="Trendline",
)


plt.title(
    "AI Replacement Risk vs. Future Demand Score", fontsize=14, fontweight="bold"
)
plt.xlabel("AI Replacement Risk (0 = Low, 1 = High)")
plt.ylabel("Future Demand Score (0 = Low, 1 = High)")
plt.legend()
plt.tight_layout()
plt.show()
markdown_text="""
this plot illustrates the relationship between automaton risk and future, and we can observe that there is a weak correlation between AI risks and Future demand. which means that a job's risk of being replaced by AI does not influence its predicted future demand 
"""
display(Markdown(markdown_text))

# %% [markdown]
# ## 6. Average Salary by Top 10 Job Titles
# Let's find out which job titles command the highest average salaries in USD.

# %%
plt.figure(figsize=(12, 6))

# Calculate average salary by job title and get top 10
top_salaries = (
    df.groupby("Job_Title")["Average_Salary_USD"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

# Horizontal bar chart
plt.barh(top_salaries.index[::-1], top_salaries.values[::-1], color="#9C27B0")
plt.title(
    "Top 10 Job Titles by Average Salary (USD)", fontsize=14, fontweight="bold"
)
markdown_text="""
"""
display(Markdown(markdown_text))
plt.xlabel("Average Salary ($)")
plt.ylabel("Job Title")
plt.tight_layout()
plt.show()
markdown_text="""
we can observe that an HR Specialist and a Cloud Architect have the highest average salaries that could reach $120000
"""
display(Markdown(markdown_text))

```
